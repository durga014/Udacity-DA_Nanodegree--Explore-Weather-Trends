```python
%matplotlib inline
%config InlineBackend.figure_format='retina'
import pandas as pd 
import matplotlib.pyplot as plt

df=pd.read_csv("D:\\Banga.csv")

mavg=df['CityAvgTemp'].rolling(10).mean()
cavag=df['GlobalAvgTemp'].rolling(10).mean()

plt.plot(df['year'],mavg,label='Global')
plt.plot(df['year'],cavag,label='city')
plt.legend()
plt.xlabel('year')
plt.ylabel('Degree')
plt.title("Bangalore City Average Temperature")
plt.show()
```
